```python
import math
def dev(x,y):
    
    return 6*x,5*math.exp(-y)

    

    
```

```python
x=4
y=3
for i in range(1000):
    del_x,del_y=dev(x,y)
    x=x-0.001*del_x
    print("x value "+str(i)+"iteration"+str(x))
    y=y-0.001*del_y
    print("y value "+str(i)+"iteration"+str(y))
    
print("final value of x ="+str(x))
print("final value of y="+str(y))
```
